```python
# 8-2 Model distributions, Continous Random Var, Probabiolity Density Function

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
from scipy import special
```

```python
# Exponential distributon
# f(x) = lambda * exp[-lambda *x] <- 이런식으로 정의됨
#PMF X -> PDF로 그려야 함
```

```python
for lam in [2.0, 1.0, 0.5]:
    x = np.linspace(0, 3, 100) #0부터 3까지 균일 간격으로 100개의 값이 들어간 어레이 생성
    y = lam * np.exp(-1 * lam *x) #정의대로 적은 식
    plt.plot(x,y, label =f"$\lambda=${lam}") #그리스어 람다로 출력
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("PDF")
    
plt.show()
plt.close()
```

```python
#PDF -> 1-e^(-lambda*x) 로 정의됨

for lam in [2.0, 1.0, 0.5]:
    x = np.linspace(0, 3, 100) #0부터 3까지 균일 간격으로 100개의 값이 들어간 어레이 생성
    y = 1 - np.exp(-1 * lam *x) #정의대로 적은 식
    plt.plot(x,y, label =f"$\lambda=${lam}") #그리스어 람다로 출력
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("CDF")
    
plt.show()
plt.close()

#CDF로 변경~
```

```python
def values2cdf(values): #값이 들어가면 CDF로 바꿔주는애
    sorted_values = sorted(values) #오른차순 정렬
    total = len(values)
    x = []
    y = []
    for i, value in enumerate(sorted_values): #enumerate : 인덱스 하나씩 붙여줌
        x.append(value)
        y.append(i/total)
    return x,y
```
