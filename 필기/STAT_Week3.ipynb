```python
#딕셔너리 만들기
# 케이스id : 표 맨 앞의 caseid (1번 사람의 답변.. 2번사람의 답변.. 이런식의 데이터임)

caseid = 10229 #10229번의 데이터를 보기 위해 변수 지정
preg_map = nsfg.MakePregMap(preg) #preg 데이터를 변수안에 담기
preg.outcome[preg_map[10229]].values #10229번의 value를 꺼내와서 확인하기

#출력된 값은 번호로 어떤 유형인지 매칭되어있다 (4 - 유산, 1 - 출산 ...)
```

```python
#★
from collections import defaultdict #딕셔너리(키,벨류) 키가 없는걸 조회하고싶다면 공백리스트를 만들어서 사용할 수 있다. 그걸 위한 라이브러리
preg_map2 = defaultdict(list) #preg_map2를 디폴트 딕셔너리로 지정함
for idx, caseid in preg.caseid.iteritems(): #표의 데이터를 한줄(아래쪽으로 한줄 한줄)씩 iteratior 하게 된다.
    preg_map2[caseid].append(idx) #caseid의 값을 돌면서 인덱스를 하나씩 추가함
    
preg.outcome[preg_map2[10229]].values
#위와 동일한 값이 나옵니다
```

```python
#태어났을 때 평균 무게 구하기
preg['totalwgt_lb'].mean() #토탈웨이트의 평균 무게 계산(산술평균) 명령어
```

```python
# pound(lb) -> kg으로 변경하고 싶은 경우!

def lb_to_kg(weightinlb):
    return 0.45359237 * weightinlb
preg["totalwgt_kg"] = preg["totalwgt_lb"].apply(lb_to_kg) #apply : 한 값을 조회해서 함수 안의 식으로 계산하여 결과값을 시리즈로 만들어준다. 즉, 새로운 칼럼을 만들어냄
```

```python
preg[["totalwgt_lb","totalwgt_kg"]] #칼럼 이름을 리스트의 요소로 주변 칼럼 안의 데이터프레임을 반환해준다
```

```python
preg["totalwgt_kg2"] = preg["totalwgt_lb"] *0.45359237
preg["totalwgt_kg2"] #위와 동일한 값 출력, 함수 없이 간단하게 가능
```

```python
preg[["totalwgt_kg", "totalwgt_kg2", "totalwgt_lb"]] #리스트의 인자로 준 세개의 칼럼의 값이 출력된다. 
```
